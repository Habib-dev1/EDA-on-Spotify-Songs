```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = 'whitegrid')
```

```python
df = pd.read_csv('spotify_dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
# 1 How many unique artists are in the dataset?
df['Artist'].nunique()
```

```python
# 2 What is the distribution of song popularity scores?
df['Popularity'] = pd.to_numeric(df['Popularity'], errors='coerce')
display(df['Popularity'].describe())
```

```python
# 3 Which artists appear most frequently in the Top 200?
df['Artist'].value_counts().head(10)
```

```python
# 4 What are the most common genres?
df['Genre'].value_counts().head(10)
```

```python
# 5 What is the average duration of songs?
df['Duration (ms)'] = pd.to_numeric(df['Duration (ms)'], errors='coerce')
df['Duration (ms)'].describe()
```

```python
# 6 How do audio features like danceability, energy, and valence vary across genres?
df_audio_features = df[['Genre', 'Danceability', 'Energy', 'Valence']]
df_audio_features
```

```python
# 7 Which countries contribute the most to the top charts?

top_chart_threshold = 20
top_chart_df = df[df['Highest Charting Position'] <= top_chart_threshold]

print(f"Number of songs in the top {top_chart_threshold} charting positions: {len(top_chart_df)}")


top_chart_genres = top_chart_df['Genre'].value_counts().head(10)
print(f"\nMost common genres in the top {top_chart_threshold} charting positions:")
display(top_chart_genres)
```

```python
# 1 Bar chart of Top 10 artists by number of chart appearances.

top_artists = df['Artist'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_artists.index, y=top_artists.values, palette='viridis')
plt.title('Top 10 Artists by Chart Appearances')
plt.xlabel('Artist')
plt.ylabel('Number of Chart Appearances')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2 Histogram of song popularity scores.

plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=20, kde=True)
plt.title('Distribution of Song Popularity Scores')
plt.xlabel('Popularity Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3 Boxplot showing distribution of danceability across genres.

df_genres = df['Genre'].apply(lambda x: x.replace("'", "").strip("[]").split(', '))
df_genres_flat = df_genres.explode()
df['First Genre'] = df_genres.apply(lambda x: x[0] if x and x[0] else 'Unknown')

# Get the top 10 genres for a clearer plot
top_genres = df['First Genre'].value_counts().head(10).index.tolist()
df_filtered_genres = df[df['First Genre'].isin(top_genres)]

plt.figure(figsize=(14, 7))
sns.boxplot(x='First Genre', y='Danceability', data=df_filtered_genres, palette='viridis')
plt.title('Distribution of Danceability Across Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Danceability')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 4 Heatmap showing correlation between audio features (danceability, energy, valence, tempo,
# loudness).

audio_features = df[['Danceability', 'Energy', 'Valence', 'Tempo', 'Loudness']].copy()
for col in audio_features.columns:
    audio_features[col] = pd.to_numeric(audio_features[col], errors='coerce')

correlation_matrix = audio_features.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Audio Features')
plt.show()
```

```python
# 5 Line chart showing trend of popularity over time.

df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
popularity_over_time = df.groupby('Release Date')['Popularity'].mean().dropna()

plt.figure(figsize=(14, 7))
sns.lineplot(x=popularity_over_time.index, y=popularity_over_time.values)
plt.title('Average Song Popularity Over Time')
plt.xlabel('Release Date')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 6 Scatter plot of energy vs loudness colored by genre.

df_genres = df['Genre'].apply(lambda x: x.replace("'", "").strip("[]").split(', '))
df['First Genre'] = df_genres.apply(lambda x: x[0] if x and x[0] else 'Unknown')

top_genres = df['First Genre'].value_counts().head(10).index.tolist()
df_filtered_genres = df[df['First Genre'].isin(top_genres)]

plt.figure(figsize=(14, 8))
sns.scatterplot(x='Energy', y='Loudness', hue='First Genre', data=df_filtered_genres, hue_order=top_genres, palette='viridis')
plt.title('Energy vs Loudness by Genre (Top 10 Genres)')
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
